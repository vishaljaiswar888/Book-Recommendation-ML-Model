### Popularity Bases Recommender System

```python
ratings_with_name = ratings.merge(books, on="ISBN")
```

```python
num_rating_df = ratings_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()

num_rating_df.rename(columns={"Book-Rating":"num_ratings"}, inplace=True)

num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby("Book-Title").mean()["Book-Rating"].reset_index()

avg_rating_df.rename(columns={"Book-Rating":"avg_ratings"}, inplace=True)

avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on="Book-Title")

popular_df
```

```python
popular_df = popular_df[popular_df["num_ratings"]>=250].sort_values("avg_ratings", ascending=False).head(50)

popular_df
```

```python
popular_df1 = popular_df.merge(books, on="Book-Title").drop_duplicates("Book-Title")[["Book-Title", "Book-Author", "Image-URL-M", "num_ratings", "avg_ratings"]]
```

### Collaborative Filter based Recommender System

```python
x = ratings_with_name.groupby("User-ID").count()["Book-Rating"] > 200
users_200 = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name["User-ID"].isin(users_200)]
```

```python
users_200.shape         # this many users gave ratings more than 200 each
```

```python
# so we have done filtering on the basis of users now let us look at books


## for users we have chosen those users who had given more than 200 ratings
```

```python
y = filtered_rating.groupby("Book-Title").count()["Book-Rating"]>50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
pt
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt) 

# finding euclideon distance for every book vector
```

```python
# for finding the index of any book

np.where(pt.index=="Zoya")[0][0]
```

```python
sorted(list(enumerate(similarity_scores[0])), key=lambda x:x[1], reverse=True)[1:6]

# enumerate: distance ke sath-sath item ka index bhi dikh jayega
```

```python
def recommend_books(book_name):
    # fetch book index
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
def recommend_books2(book_name):
    # fetch book index
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:6]
    
    data=[]
    for i in similar_items:
        item = []
        
        # list ko append karne se 2D list ban jata hai, that's why extend
        temp_df = books[books["Book-Title"]==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))
        
        data.append(item)
    
    return data
```
